```python
import pandas as pd
from geopy.geocoders import Nominatim
import urllib.request, urllib.parse, urllib.error
import os, json, requests
import json
import ssl

df = pd.read_csv('cidades_regiao_cafe.csv', encoding ='latin1')
df
```

```python
def get_estado(uf):
    estados = {
        'AC': 'Acre',
        'AL':'Alagoas',
        'AP':'Amapá',
        'AM':'Amazonas',
        'BA':'Bahia',
        'CE':'Ceará',
        'ES':'Espírito Santo',
        'GO':'Goiás',
        'MA':'Maranhão',
        'MT':'Mato Grosso',
        'MS':'Mato Grosso do Sul',
        'MG':'Minas Gerais',
        'PA':'Pará',
        'PB':'Paraíba',
        'PR':'Paraná',
        'PE': 'Pernambuco',
        'PI':'Piauí',
        'RJ':'Rio de Janeiro',
        'RN':'Rio Grande do Norte',
        'RS':'Rio Grande do Sul',
        'RO':'Rondônia',
        'RR':'Roraima',
        'SC':'Santa Catarina',
        'SP':'São Paulo',
        'SE':'Sergipe',
        'TO': 'Tocantins',
        'DF': 'Distrito Federal'
            }
    return estados[uf]

df['Estado'] = df.apply(lambda x: get_estado(x.loc['UF']), axis=1).copy()
df
```

```python
df['location'] = df['Município'] + ", " + df['UF'] + ", Brazil"
df.head()
```

```python
def get_gps(address):
    geolocator = Nominatim(user_agent="eu")
    location = geolocator.geocode(address)
    try:
        lat = location.latitude
        long = location.longitude
        cidade = location.address.split(',')[0]
        ready = str(lat) +","+ str(long)+"," + cidade
        print(ready)
        return ready
    except:
        return "None"
df['gps'] = df.apply(lambda x: get_gps(x.loc['location']), axis=1)
df
```
